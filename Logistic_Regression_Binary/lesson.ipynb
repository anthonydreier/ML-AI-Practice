```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
plt.scatter(df['age'], df.bought_insurance, marker='x', color='black')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, train_size=0.9, random_state=42)
```

```python
X_test
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
plt.scatter(df['age'], df.bought_insurance, marker='x', color='black')
plt.scatter(X_test, model.predict(X_test), color='red')
```

```python
model.score(X_test, y_test)
```

```python
model.predict_proba(X_test)
```
